+ Integral Numerik Simpson 1/3 Komposit

Volume benda putar terhadap sumbu-$x$ diberikan oleh

$$
V = \pi \int_a^b [f(x)]^2 \, dx
$$

Untuk menghitung integral secara numerik, digunakan metode Trapesium Komposit dan Simpson 1/3 Komposit.
+ Metode Trapesium Komposit

Rumus integral komposit metode trapesium adalah

$$
\int_a^b f(x)\,dx 
\approx 
\frac{h}{2}
\left[
f(x_0)
+
2 \sum_{i=1}^{n-1} f(x_i)
+
f(x_n)
\right]
$$

dengan

$$
h = \frac{b-a}{n}
$$

Jika diterapkan pada volume benda putar, maka

$$
V \approx 
\frac{h}{2}
\left[
g(x_0)
+
2 \sum_{i=1}^{n-1} g(x_i)
+
g(x_n)
\right]
$$

dengan

$$
g(x) = \pi [f(x)]^2.
$$


+ Metode Simpson 1/3 Komposit

Metode Simpson 1/3 Komposit mensyaratkan jumlah partisi $n$ harus genap.

Rumusnya adalah

$$
\int_a^b f(x)\,dx 
\approx 
\frac{h}{3}
\left[
f(x_0)
+
4 \sum_{\substack{i=1 \\ i \text{ ganjil}}}^{n-1} f(x_i)
+
2 \sum_{\substack{i=2 \\ i \text{ genap}}}^{n-2} f(x_i)
+
f(x_n)
\right]
$$

Sehingga untuk volume benda putar diperoleh

$$
V
\approx
\frac{h}{3}
\left[
g(x_0)
+
4 \sum_{\substack{i=1 \\ i \text{ ganjil}}}^{n-1} g(x_i)
+
2 \sum_{\substack{i=2 \\ i \text{ genap}}}^{n-2} g(x_i)
+
g(x_n)
\right]
$$


+ Contoh Kasus

Misalkan $f(x) = x^2$ pada interval $[0,1]$.

Maka volume eksaknya adalah

$$
V = \pi \int_0^1 x^4 \, dx
$$

$$
V = \pi \left[ \frac{x^5}{5} \right]_0^1
$$

$$
V = \frac{\pi}{5}.
$$

```python
import numpy as np

# Fungsi asli
def f(x):
    return x**2

# Fungsi untuk volume benda putar
def g(x):
    return np.pi * (f(x))**2


# -------------------------------
# METODE TRAPESIUM KOMPOSIT
# -------------------------------
def trapezoidal(a, b, n):
    h = (b - a) / n
    x = np.linspace(a, b, n+1)
    
    total = g(x[0]) + g(x[-1])
    
    for i in range(1, n):
        total += 2 * g(x[i])
    
    return (h/2) * total


# -------------------------------
# METODE SIMPSON 1/3 KOMPOSIT
# -------------------------------
def simpson(a, b, n):
    # cek apakah n genap
    if n % 2 != 0:
        raise ValueError("n harus genap untuk Simpson 1/3")
    
    h = (b - a) / n
    x = np.linspace(a, b, n+1)
    
    total = g(x[0]) + g(x[-1])
    
    for i in range(1, n):
        if i % 2 == 0:
            total += 2 * g(x[i])
        else:
            total += 4 * g(x[i])
    
    return (h/3) * total


# --------------------------------
# MAIN PROGRAM
# --------------------------------
a = 0
b = 1
n = 10   # harus genap untuk Simpson

V_trap = trapezoidal(a, b, n)
V_simp = simpson(a, b, n)

print("Volume (Trapesium) =", V_trap)
print("Volume (Simpson)   =", V_simp)
print("Selisih =", abs(V_simp - V_trap))
```
